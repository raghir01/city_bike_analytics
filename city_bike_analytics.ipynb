```python
resource_path = './Resources'
only_files = [join(resource_path, f) for f in os.listdir(resource_path) if isfile(join(resource_path, f)) 
              and f.find('.csv') != -1]
dfs = []
for f in only_files:
    df = pd.read_csv(f)
    dfs.append(df)
citibike_2018 = pd.concat(dfs)

citibike_2018.head()
citibike_2018.to_csv('citibike_2018_data.csv', sep = ',')
citibike_2018.head()
```
